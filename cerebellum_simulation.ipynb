```python
from src.mf_definition import cerebellar_mean_field
from scipy.optimize import fsolve

# Initial guess for firing rates [GrC, GoC, PC, MLI] (Hz)
nu_initial_guess = [80.0, 30.0, 40.0, 0.0]

# External input rate (e.g., mossy fiber input to GrC)
mf_input_rate = 800.0  # Hz
cf_input_rate = 0 # Hz

# Solve for steady-state rates
nu_steady = fsolve(
    cerebellar_mean_field, 
    nu_initial_guess, 
    args=(mf_input_rate, cf_input_rate)
)

print(f"Steady-State Rates:")
print(f"GrC: {nu_steady[0]:.2f} Hz, GoC: {nu_steady[1]:.2f} Hz")
print(f"PC: {nu_steady[2]:.2f} Hz, MLI: {nu_steady[3]:.2f} Hz")
```
